# **Test Model Accuracy**

```python
import os
import json
import numpy as np
from PIL import Image
import torch
from ultralytics import YOLO
from collections import Counter, defaultdict
import pandas as pd

def process_test_data(model_path, test_dir, output_json='class_matrix.json', output_csv='best_class_matrix_.csv', img_size=256):
    """
    Process test data and generate a confusion matrix and class statistics

    Args:
        model_path: Path to the YOLO model
        test_dir: Directory containing folders of classes with test images
        output_json: Path to save the output JSON
        output_csv: Path to save the output CSV
        img_size: Size to resize images to
    """
    # Load the model
    print(f"Loading model from {model_path}...")
    model = YOLO(model_path)

    # Get class names from folder names in test_dir
    class_names = [folder for folder in os.listdir(test_dir)
                  if os.path.isdir(os.path.join(test_dir, folder))]
    class_names.sort()  # Ensure consistent ordering

    print(f"Found {len(class_names)} classes: {class_names}")

    # Create confusion matrix
    num_classes = len(class_names)
    confusion_matrix = np.zeros((num_classes, num_classes), dtype=int)

    # Dictionary to map class names to their indices
    class_to_idx = {name: idx for idx, name in enumerate(class_names)}

    # For tracking per-class statistics
    class_stats = []

    # Process each class folder
    for true_class in class_names:
        class_path = os.path.join(test_dir, true_class)
        true_idx = class_to_idx[true_class]

        # Get valid image files
        valid_extensions = ['.jpg', '.jpeg', '.png', '.bmp']
        image_files = [f for f in os.listdir(class_path)
                      if os.path.isfile(os.path.join(class_path, f)) and
                      any(f.lower().endswith(ext) for ext in valid_extensions)]

        if not image_files:
            print(f"No images found for class {true_class}. Skipping.")
            continue

        total_samples = len(image_files)
        print(f"Processing {total_samples} images for class '{true_class}'...")

        # To track what classes the current class gets confused with
        confused_with = Counter()

        # Process each image
        for img_file in image_files:
            img_path = os.path.join(class_path, img_file)

            try:
                # Open and resize image
                img = Image.open(img_path).convert('RGB')
                img = img.resize((img_size, img_size))

                # Get prediction
                results = model(img)
                probs = results[0].probs  # Get probability distribution

                # Get top prediction
                pred_idx = int(probs.top1)  # Index of the top prediction
                pred_class = class_names[pred_idx]

                # Update confusion matrix
                confusion_matrix[true_idx][pred_idx] += 1

                # If prediction is wrong, update confused_with counter
                if pred_idx != true_idx:
                    confused_with[pred_class] += 1

            except Exception as e:
                print(f"Error processing {img_path}: {e}")

        # Calculate metrics for this class
        true_pos = confusion_matrix[true_idx][true_idx]
        false_neg = total_samples - true_pos
        false_pos = sum(confusion_matrix[j][true_idx] for j in range(num_classes) if j != true_idx)

        # Calculate accuracy and precision
        accuracy = true_pos / total_samples if total_samples > 0 else 0
        precision = true_pos / (true_pos + false_pos) if (true_pos + false_pos) > 0 else 0

        # Get top 3 confused classes
        most_confused = ["None", "None", "None"]
        confused_counts = [0, 0, 0]

        for idx, (confused_class, count) in enumerate(confused_with.most_common(3)):
            if idx < 3:
                most_confused[idx] = confused_class
                confused_counts[idx] = count

        # Create class statistics entry
        class_stat = {
            'Class_ID': int(true_idx),
            'Class_Name': true_class,
            'Total_Samples': int(total_samples),
            'Correct_Predictions': int(true_pos),
            'False_Positives': int(false_pos),
            'False_Negatives': int(false_neg),
            'Accuracy': float(accuracy),
            'Precision': float(precision),
            'Most_Confused_With_Class1': most_confused[0],
            'Confused_Count1': int(confused_counts[0]),
            'Most_Confused_With_Class2': most_confused[1],
            'Confused_Count2': int(confused_counts[1]),
            'Most_Confused_With_Class3': most_confused[2],
            'Confused_Count3': int(confused_counts[2]),
            'Accuracy_Percentage': float(accuracy * 100),
            'Precision_Percentage': float(precision * 100)
        }

        class_stats.append(class_stat)
        print(f"Completed class '{true_class}' with accuracy: {accuracy:.2%}")

    # Calculate overall metrics
    overall_accuracy = np.trace(confusion_matrix) / np.sum(confusion_matrix) if np.sum(confusion_matrix) > 0 else 0

    # Create results dictionary
    results = {
        'overall_accuracy': float(overall_accuracy),
        'overall_accuracy_percentage': float(overall_accuracy * 100),
        'confusion_matrix': confusion_matrix.tolist(),
        'class_names': class_names,
        'class_statistics': class_stats
    }

    # Convert NumPy types to Python native types for JSON serialization
    def convert_to_serializable(obj):
        if isinstance(obj, dict):
            return {k: convert_to_serializable(v) for k, v in obj.items()}
        elif isinstance(obj, list):
            return [convert_to_serializable(i) for i in obj]
        elif isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        else:
            return obj

    # Convert all values to JSON serializable types
    results = convert_to_serializable(results)

    # Save results to JSON file
    with open(output_json, 'w') as f:
        json.dump(results, f, indent=4)

    # Save class statistics to CSV file
    df = pd.DataFrame(class_stats)
    df.to_csv(output_csv, index=False)

    print(f"\nResults saved to {output_json} and {output_csv}")
    print(f"Overall accuracy: {overall_accuracy:.2%}")

    # Create a dataframe for better viewing
    df = pd.DataFrame(class_stats)
    print("\nClass Statistics Summary:")
    print(df[['Class_Name', 'Total_Samples', 'Accuracy_Percentage', 'Precision_Percentage']])

    return results

if __name__ == "__main__":
    # Example usage
    model_path = "/content/drive/MyDrive/yolov11x 100 class 256x256 dataset 3 Agumented/weights/best.pt"  # Update this path as needed
    test_dir = "/content/flower_dataset/content/randomly_split_dataset_512x512/test"  # Update with your test directory

    # Get the test directory from user input if not specified
    if test_dir == "path/to/test/folder":
        test_dir = input("Please enter the path to your test directory: ")

    # Process the test data
    results = process_test_data(model_path, test_dir)
```
